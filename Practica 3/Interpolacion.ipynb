Ejercicio: Interpolar con 10 puntos generados de la función de Runge $f(x) = \frac{1}{1+25x^2}$ en el intervalo $[-1,1]$ equidistantes. Hacer la gráfica con $10,000$ puntos en el mismo intervalo. Utilizar polyfit para el polinomio interpolador y splines cúbicos.

```python
#Importamos librerias
import numpy as np
import matplotlib.pyplot as plt
import pprint
```

```python
#Definimos la funcion Runge
def f_runge(x):
    return 1/(1+25*(x**2))
```

```python
x=np.linspace(-1, 1,npoints) #espacio de x del intervalo -1 a 1 con incrementos del número de puntos que deseamos
y=f_runge(x)
```

```python
npoints=10
ngrado=npoints-1
points=np.polyfit(x,y,ngrado)
pprint.pprint(points)
np.set_printoptions(precision = 2) #Para ajustar a 2 decimales
```

```python
plt.plot(x,y,'o')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Puntos Runge')
```

```python
#Grafica-Interpolación
n_peval=10000
x_eval=np.linspace(min(x),max(x),n_peval)
y_eval=np.polyval(points,x_eval)
```

```python
plt.plot(x,y,'o',x_eval,y_eval,'k-')
plt.legend(['datos','interpolacion'])
```

```python
from scipy.interpolate import interp1d
```

```python
#splineas cúbicos
spline=interp1d(x,y,kind='cubic')
n_peval=10000
x_eval=np.linspace(min(x),max(x),n_peval)
y_eval=spline(x_eval)
```

```python
plt.plot(x,y,'o',x_eval,y_eval,'k-')
plt.legend(['datos','spline'])
```
