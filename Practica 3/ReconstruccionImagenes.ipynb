## Aplicación: reconstrucción de imágenes
Dentro de las aplicaciones de la SVD de una matriz se encuentra la de reconstrucción de imágenes.

Usando 10, 20, 30, 40 y 50 número de vectores singulares izquierdos, derechos y valores singulares¶

```python
#Importamos Librerias
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
#Leemos la imagen
img=plt.imread('peak.png',0)
```

```python
img
```

```python
#Canales
img.shape
img[:,:,0].shape
img = img[:,:,0]
```

```python
imgplot = plt.imshow(img,cmap='gray')
plt.title('Foto en grises')
```

```python
#SVD a la imagen
U,S,V_T = np.linalg.svd(img, full_matrices=False)
```

```python
print(U.shape)
print(S.shape)
print(V_T.shape)
```

```python
img_svd = (U*S)@V_T
img_svd.shape
```

```python
plt.imshow(img_svd, cmap='gray')
plt.title('Foto con SVD')
```

Usando 10, 20, 30, 40 y 50 número de vectores singulares izquierdos, derechos y valores singulares


```python
for i in [10, 20, 30, 40]:
    imag=np.matrix(U[:, :i]) * np.diag(S[:i]) * np.matrix(V_T[:i, :])
    plt.imshow(imag, cmap='gray')
    title = "n = %s" % i
    plt.title(title)
    plt.show()
```
